```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

1. Linear regression function

```python
# Add data
n = int(input("Amount of pairs of data:"))
X = []
Y = []
 
for i in range(n):
    x_data = float(input("Input the value of x: "))
    y_data = float(input("Input the value of y: "))
    X.append(x_data)
    Y.append(y_data)
    
print(X)
print(Y)
```

```python
# Linear Regression
c_0 = 0
c_1 = 0
Z = np.multiply(X,Y)
x = 0

def linear_regression():
    x_sum = sum(X)
    y_sum = sum(Y)
    xy_sum = sum(Z)
    x_ssum = sum(i*i for i in X)
    y_ssum = sum(j*j for j in Y)
    c_0 = (x_ssum*y_sum - x_sum*xy_sum)/(n*x_ssum - x_sum**2)
    c_1 = (n*xy_sum - x_sum*y_sum)/(n*x_ssum - x_sum**2)
    x = float(input("Choose x whose y value is to be evaluated: "))
    return c_0 + c_1*x
    
linear_regression()
```

2. Linear Interpolation

```python
# Linear Interpolation

def linear_interpol():
    y = np.interp(float(input("Choose x whose y is to be evaluated: ")), X, Y)
    print(y)
    
linear_interpol()
```

3. Lagrange Interpolation

```python
# Lagrange Interpolation
from scipy.interpolate import lagrange

def lagrange_interpol():
    f = lagrange(X, Y)
    print(f(float(input("Choose x whose y is to be evaluated: "))))
    
lagrange_interpol()
```

4. Plots

```python
# All functions are redefined as they are used for different purpose
# Lagrange interpolation

def lagrange_interpol(x):
    g = lagrange(X, Y)
    return g(x)
    
x_new = np.arange(-0.0, 3.1, 0.1)
fig = plt.figure(figsize = (10,8))
plt.plot(x_new, lagrange_interpol(x_new), 'b', X, Y, 'ro')
plt.title('Lagrange Polynomial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()    
```

```python
# Linear interpolation
from scipy.interpolate import interp1d

def linear_interpol(x):
    h = interp1d(X, Y)
    return h(x)
    
x_new = np.arange(0.0, 3.1, 0.1)
fig = plt.figure(figsize = (10,8))
plt.plot(x_new, linear_interpol(x_new), 'b', X, Y, 'ro')
plt.title('Linear Interpolation')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()    
```
