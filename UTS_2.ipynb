2. Write linear equations for $I_1$, $I_2$, $I_3$, and $I_4$ from produced figure.

Using KVL:
$$\epsilon_1 = I_1 R_1 + (I_1 + I_2) R_5 + (I_1 + I_3) R_6$$
$$\epsilon_2 = I_2 R_2 + (I_1 + I_2) R_5 + (I_2 + I_4) R_8$$
$$\epsilon_3 = I_3 R_3 + (I_3 + I_4) R_7 + (I_1 + I_3) R_6$$
$$\epsilon_4 = I_4 R_4 + (I_3 + I_4) R_7 + (I_2 + I_4) R_8$$

The resulting equation is
$$\epsilon_1 = (R_1 + R_5 + R_6)I_1 + R_5 I_2 + R_6 I_3 + (0) I_4$$
$$\epsilon_2 = R_5 I_1 + (R_2 + R_5 + R_8) I_2 + (0) I_3 + R_8 I_4$$
$$\epsilon_3 = R_6 I_1 + (0) I_2 + (R_3 + R_6 + R_7) I_3 + R_7 I_4$$
$$\epsilon_4 = (0) I_1 + R_8 I_2 + R_7 I_3 + (R_4 + R_7 + R_8) I_4$$
