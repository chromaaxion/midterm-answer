1. Numerical integration of $f(x) = \sin(x)$

```python
# Define function

def f(x):
    return math.sin(x)

xbeg = float(input("Input xbeg: "))
xend = float(input("Input xend: "))
N = int(input("Input partition number: "))
dx = (xend - xbeg)/N

def rectl():
    x = xbeg
    area = 0
    
    for i in range(N):
        area += f(x) * dx
        x += dx
        
    return area

def rectr():
    x = dx
    area = 0
    
    for i in range(N):
        area += f(x) * dx
        x += dx
        
    return area

def trapez():
    x = xbeg
    area = 0
    
    for i in range(N):
        area += (dx/2) * (f(x) * f(x + dx))
        x += dx
        
    return area

def simpson():
    x = xbeg
    area = 0
    
    for i in range(N):
        area += (dx/6) * (f(x) + 4*f((x + dx)/2) + f(x + dx))
        x += dx
    
    return area

rectl()
```

2. Finite Difference Method

```python
# n = 10

import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline

g = 10

n = 10
h = (5-0) / n

# Get A
A = np.zeros((n+1, n+1))
A[0, 0] = 1
A[n, n] = 1
for i in range(1, n):
    A[i, i-1] = 1
    A[i, i] = -2
    A[i, i+1] = 1

print(A)

# Get b

b = np.zeros(n+1)
b[0] = 10
b[1:-1] = -g*h**2
b[-1] = 20
print(b)

# solve the linear equations
y = np.linalg.solve(A, b)

t = np.linspace(0, 5, 11)

plt.figure(figsize=(10,8))
plt.plot(t, y)
plt.plot(5, 20, 'ro')
plt.xlabel('time (s)')
plt.ylabel('altitude (m)')
plt.show()
```
